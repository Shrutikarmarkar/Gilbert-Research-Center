```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
# x = minutes
x = [1,2,6,7,8,9,11,12,13,14,15,16,17,18,21,22]

# y = recorded number of E-coli bacteria cells
y = []

for minutes in x:
  bacteria = 2 ** minutes;   # the equation y = 2 ** x from handwritten solution shows exponential growth of E-Coli Bacteria cells
  y.append(bacteria)
```

```python
number_of_bins = int((math.log(len(x),2))+1 )   # Strunge's Formula
number_of_bins
```

```python
print(y)
```

```python
plt.hist(y, bins=number_of_bins, weights=np.ones_like(y)/len(y), edgecolor='black', alpha=0.7)
plt.xlabel('Number of E-coli Bacteria Cells')
plt.ylabel('Relative Frequency of E-coli Bacteria Cells')
plt.title('Histogram of E-coli Bacteria Cell plotted in relative frequency')
plt.show()
```

The histogram shows relative frequency of the recorded number of E-coli bacteria cells. Plotting a histogram in terms of relative frequency means that the height of each bar represents the proportion of data points that fall within that bin, rather than the actual count of data points.

From the above Histogram it is clear that with time the number of bacteria grew exponentially, therefore, most of the recorded E-Coli bacteria are below 10000
