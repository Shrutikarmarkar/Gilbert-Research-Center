```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')

z = np.linspace(0, 20, 100)  # specifying the z axis with different continuous values of t
x = 2 * np.sqrt(2) * np.sin(z)
y = 2 * np.sqrt(2) * np.cos(z)

ax.plot3D(x, y, z, 'blue')
ax.set_title('RNA SARS-CoV-2 structure')

ax.set_xlabel('x(t)')
ax.set_ylabel('y(t)')
ax.set_zlabel('z(t)')
plt.show()
```
